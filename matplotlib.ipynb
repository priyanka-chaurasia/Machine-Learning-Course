```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline    
```

```python
import numpy as np
x = np.linspace(1,5,11)
y =  x**2
x
```

```python
y
```

```python
#functional method
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Title Here')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
plt.subplot(1,2,2)
plt.plot(x,y,'g*-')
```

```python
#object oriented
fig = plt.figure()

axes1= fig.add_axes([0.1,0.1,0.9,0.9])
axes1.plot(x,y)
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('Large')

axes2 = fig.add_axes([0.25,0.6,0.4,0.2])
axes2.plot(y,x)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Small')
```

```python
fig,axes = plt.subplots()
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize=(10,3))
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3)
axes
plt.tight_layout()
```

```python
for ax in axes:
    ax.plot(x,y)
    ax.set_xlabel("X ")
    ax.set_ylabel('Y')
    ax.set_title('Title')
plt.tight_layout()
fig
```

```python
fig.savefig('Matplotlib-practice.png')
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([1,1,1,1])
ax.plot(x,y**2,label="y squared")
ax.plot(x,y*4,label="y multiplied by 4")
ax.legend()
ax.legend(loc = 10)

```

```python
fig1 = plt.figure()
ax = fig1.add_axes([1,1,1,1])
ax.plot(x,y,color="purple",linewidth="2",marker="o")
```
